```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
```

```python
properties=pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col=None)
```

```python
properties.head()
```

```python
properties=properties.transpose()
```

```python
properties.head()
```

```python
properties=properties.reset_index()
```

```python
properties.head()
```

```python
new_header = properties.iloc[0] #grab the first row for the header
properties = properties[1:] #take the data less the header row
properties.columns = new_header
```

```python
properties.head(5)
```

```python
properties.columns.values[0] = "borough"
properties.columns.values[1] = "borough_key"
properties.head()
```

```python
properties=properties.iloc[:,:317]
```

```python
properties_melt=properties.melt(id_vars=['borough','borough_key'],var_name='time',value_name='price')
properties_melt.head()
```

```python
properties_melt['price']= properties_melt['price'].astype(float) 
print(properties_melt.dtypes)
```

```python
properties_melt.shape
```

```python
properties_melt.isna().any()
```

```python
properties_exc_na = properties_melt[properties_melt['price'].notna()]
properties_exc_na.shape
```

```python
properties_exc_na.isna().any()
```

```python
properties_exc_na['year']=pd.DatetimeIndex(properties_exc_na['time']).year
properties_exc_na.head()
```

```python
prp_yrly=properties_exc_na[['borough','year','price']]
prp_yrly
```

```python
properties_yrly_mean=prp_yrly.groupby(['borough','year']).mean()
properties_yrly_mean
```

```python
prp_y=properties_yrly_mean.reset_index()
prp_y.head()
```

```python
prp_1998 = prp_y[prp_y['year']==1998]
prp_1998.head()
```

```python
prp_2018 = prp_y[prp_y['year']==2018]
prp_2018.head()
```

```python
prp_1998.count()
```

```python
Brent_1998 = float(prp_1998['price'][prp_1998['borough'] == 'Brent'])
Brent_1998
```

```python
Brent_2018 = float(prp_2018['price'][prp_2018['borough'] == 'Brent'])
Brent_2018
```

```python
Brent_2018/Brent_1998
```

```python
def price_ratio (x):
    prp_1998_p = float(prp_1998['price'][prp_1998['borough']== x])
    prp_2018_p = float(prp_2018['price'][prp_2018['borough']== x])
    ratio = prp_2018_p/prp_1998_p
    return ratio
```

```python
price_ratio('Brent')
```

```python
all_borough=prp_2018['borough'].unique() 
all_borough
```

```python
## The following code can only provide a list of average values 
## ==> we need a dictionary which contains {borough:meanPrice}
##for i in range(len(all_borough)):  
##    print(price_ratio(all_borough[i]))
```

```python
br_avg_p={}
for br in all_borough: 
    all_br_pr= price_ratio(br)
    br_avg_p[br]= all_br_pr
br_avg_p  
```

```python
all_br_ratio = pd.DataFrame(br_avg_p.items(),columns = ['borough', 'ratio'])
all_br_ratio
```

```python
top5_avg_inc=all_br_ratio.sort_values(by='ratio', ascending = False).head()
top5_avg_inc
```

```python
import seaborn as sns
ax = sns.barplot(x="borough", y="ratio", data=top5_avg_inc)
```
